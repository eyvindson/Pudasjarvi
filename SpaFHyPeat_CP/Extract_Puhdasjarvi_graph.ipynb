```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

for part in ["_data_51","_data_52","_data_53"]:
    folder = "/scratch/project_2003225/GIT/KE/SpaFHyPeat_CP/Puhdasjarvi"+part+"/"

    Data = {}
    for i in range(2020,2061):#[2020,2021,2022,2023,2024,2025,2026,2027,2028,2029,2031,2032,2033,2034,2035,2036,2037,2038,2039,2041,2042,2043,2044,2045,2046,2047,2048,2049]:
        Data[i] = pd.read_csv(folder+"Puhdasjarvi"+part+"EMISSIONS_output"+str(i)+".csv")

    t3 = {}
    t3["Segment"] = Data[2020]["ORIG_seg"]
    t3["Kasvu"] = Data[2020]["Kasvu"]

    t3["CO2_sum_yearly_mean"] = sum([Data[i][["CO2"+str(ii) for ii in range(2000,2015)]].mean(axis=1) for i in range(2020,2061)])/41
    t3["N2O_sum_yearly_mean"] = sum([Data[i][["N2O"+str(ii) for ii in range(2000,2015)]].mean(axis=1) for i in range(2020,2061)])/41
    t3["CH4_sum_yearly_mean"] = sum([Data[i][["CH4"+str(ii) for ii in range(2000,2015)]].mean(axis=1) for i in range(2020,2061)])/41
    t3["CO2_eqv_sum_yearly_mean"] = sum([Data[i][["CO2"+str(ii) for ii in range(2000,2015)]].mean(axis=1)+ Data[i][["N2O"+str(ii) for ii in range(2000,2015)]].mean(axis=1)*298+Data[i][["CH4"+str(ii) for ii in range(2000,2015)]].mean(axis=1)*84  for i in range(2020,2061)])/41
    t3["Osuus"] = Data[2020]["Osuus"]        
    dat = pd.DataFrame(t3)
    dat.to_csv("/scratch/project_2003225/GIT/KE/SpaFHyPeat_CP/results/output"+part+"_total.csv")
    fig, (ax1,ax2,ax3,ax4,ax5) = plt.subplots(5,1,sharex="col",sharey="col")

    l1 = dat[dat["Kasvu"].isin([1,2])]["CO2_eqv_sum_yearly_mean"].plot.hist(bins=60,ax = ax1,title = "Kasvu_1 & 2")
    l2 = dat[dat["Kasvu"] == 3]["CO2_eqv_sum_yearly_mean"].plot.hist(bins=60,ax = ax2,title = "Kasvu_3")
    l3 = dat[dat["Kasvu"] == 4]["CO2_eqv_sum_yearly_mean"].plot.hist(bins=60,ax = ax3,title = "Kasvu_4")
    l4 = dat[dat["Kasvu"] == 5]["CO2_eqv_sum_yearly_mean"].plot.hist(bins=60,ax = ax4,title = "Kasvu_5")
    l5 = dat[dat["Kasvu"] == 6]["CO2_eqv_sum_yearly_mean"].plot.hist(bins=60,ax = ax5,title = "Kasvu_6")
    ax5.set_xlabel("g/ $m^2$")
    fig.tight_layout(pad=1.0,rect=(-.75,-.5,2,2))
    plt.savefig("/scratch/project_2003225/GIT/KE/SpaFHyPeat_CP/results/kasvu_CO2_Eqv"+part+".png",bbox_inches = "tight")     
```
